```python
import pandas as pd
```

```python
# Intialize dictionary to store CRIDs from each dataset
CRID_unique = {}
```

# May FOIA

```python
# Read FOIA_May dataset and print sheetnames
may_xls = pd.ExcelFile('Raw/raw dump May2016/IPRA Shootings Data.xls')
may_xls.sheet_names
```

```python
# Year and corresponding sheetname
may_sheets = {
    '2016' : 'qry_II_217_Data163_2016',
    '2015' : 'qry_II_217_Data163_2015',
    '2014' : 'qry_II_217_Data163_2014',
    '2013' : 'qry_II_217_Data163_2013',
    '2012' : 'qry_II_217_2012_Incid',
    '2011' : 'qry_II_217_2011_Incid',
    '2010' : 'qry_II_217_2010_Incid',
    '2009' : 'qry_II_217_2009_Incid',
    '2008' : 'qry_II_217_200807_Incid',
    }
```

```python
# Extract unique CRIDs for each year and put in master dict
CRID_unique['May FOIA'] = {}
for year, sheetname in may_sheets.items():
    CRID_unique['May FOIA'][year] = [
        str(int(x)) for x in 
        may_xls.parse(sheetname)['Complaint_Number'].dropna().unique().tolist()
        ]
```

# April FOIA

```python
april_sheets = {
    '2016' : 'All complaints during 2016YTD.xls',
    '2015' : 'All complaints during 2015.xls',
    '2014' : 'All complaints during 2014.xls',
    '2013' : 'All complaints during 2013.xls',
    '2012' : 'All complaints during 2012.xls'
}
```

```python
pd.read_excel('Raw/raw dump April2016/All complaints during 2016YTD.xls').head()
```

```python
CRID_unique['April FOIA'] = {}
for year, filename in april_sheets.items():
    CRID_unique['April FOIA'][year] = [
        str(int(x)) for x in 
        pd.read_excel('Raw/raw dump April2016/' + filename)['Complaint_Number'].dropna().unique().tolist()
        ]
```

# February FOIA

```python
feb_sheets = {
    '18A' : '18A - 1 - Incident(1).xls',
    '18B' : '18b - 1 - Incidents(1).xls',
    '20A' : '20A - 1 - Incident(1).xls'
}
```

```python
pd.read_excel('Raw/raw dump Feburary2016/18A - 1 - Incident(1).xls',skiprows=range(6)).head()
```

```python
CRID_unique['February FOIA'] = {'2006-2015' : []}
for cat, filename in feb_sheets.items():
    CRID_unique['February FOIA']['2006-2015'].extend([
            str(int(x)) for x in
            pd.read_excel('Raw/raw dump Feburary2016/' + filename,skiprows=range(6))['Log No'].dropna().unique().tolist()
        ])
```

# June

# Count records

```python
for dataset, yeardict in CRID_unique.items():
    for year, CRID_list in yeardict.items():
        print(dataset,year,len(CRID_list))
```

```python
for dataset, yeardict in CRID_unique.items():
    print(dataset,sum([len(CRID_list) for CRID_list in yeardict.values()]))
```
